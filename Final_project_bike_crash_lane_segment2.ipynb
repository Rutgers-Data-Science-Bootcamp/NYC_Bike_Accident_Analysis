```python
# Load the cleaned NYC_Crash_Cyclist_Clean_2020_2022.csv dataset.
file_path = "Resources/Segment2/Final_project_NYC_Crash_Cyclist_Cleaned.csv"
crash_df = pd.read_csv(file_path, index_col=0 )
print(crash_df.shape)
crash_df.head(10)
```

```python
# clean space in the rows valus and create new column name street
def clean_space(test):
    return test.strip()
crash_df["STREET"]=crash_df["ON STREET NAME"].apply(clean_space)
crash_df1 = crash_df.drop(["ON STREET NAME"], axis=1)
```

```python
# making new dataframe by grouing borough
crash_df_borough = crash_df1.groupby(["BOROUGH"]).count()
```

```python
crash_df_borough.head()
```

```python
# making new dataframe by grouing street
crash_df_street = crash_df1.groupby(["STREET"]).count()
crash_df_street
```

```python
# reset index and make street as a column in a new dataframe
crash_df_street1=crash_df_street.reset_index().drop(["index"], axis=1)
crash_df_street1
```

```python
# realized that STREET column has some lowercase values, convert them into uppercase
crash_df_street1['STREET'] = crash_df_street1['STREET'].str.upper()
```
