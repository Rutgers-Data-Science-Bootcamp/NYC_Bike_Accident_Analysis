```python
# Initial imports
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the crypto_data.csv dataset.
file_path = "Resources/NYC_Crash_Cyclist_Clean_2020_2022.csv"
crash_df = pd.read_csv(file_path, index_col=0 )
print(crash_df.shape)
crash_df.head(10)
```

```python
def clean_space(test):
    return test.strip()
crash_df["STREET"]=crash_df["ON STREET NAME"].apply(clean_space)
crash_df1 = crash_df.drop(["ON STREET NAME"], axis=1)
```

```python
crash_df_borough = crash_df1.groupby(["BOROUGH"]).count()
```

```python
crash_df_borough.head()
```

```python
crash_df_street = crash_df1.groupby(["STREET"]).count()
```

```python
crash_df_street1=crash_df_street.reset_index().drop(["index"], axis=1)
crash_df_street1
```

```python
crash_df_street1['STREET'] = crash_df_street1['STREET'].str.upper()
```
