```python
# Load the crypto_data.csv dataset.
file_path = "Crash-Cyclist-2020.csv"
crash_df = pd.read_csv(file_path, index_col=0 )
print(crash_df.shape)
crash_df.head(10)
```

```python
# check the dataset
crash_df.info()
```

```python
# count the null value in the dataframe using for loop
for column in crash_df.columns:
    print(f"Clumn {column} has {crash_df[column].isnull().sum()}")
```

```python
# count the nan value in the dataframe using for loop which is basically same as null value counts
for column in crash_df.columns:
     print(f"Clumn {column} has {crash_df[column].isna().sum()}")
```

```python
# clean the dataframe by removing columns missing most of the values
crash_df_cleaned = crash_df.drop(["OFF STREET NAME", "CROSS STREET NAME", "CONTRIBUTING FACTOR VEHICLE 3", "CONTRIBUTING FACTOR VEHICLE 4", 
                          "CONTRIBUTING FACTOR VEHICLE 5", "VEHICLE TYPE CODE 3", "VEHICLE TYPE CODE 4", "VEHICLE TYPE CODE 5" ], axis =1)
print(crash_df_cleaned.shape)
crash_df_cleaned
```

```python
# make numerical index to the dataframe 
crash_df_cleaned.reset_index(inplace=True)
crash_df_cleaned.head(10)
```

```python
# make a new dataframe with location columns from the original dataframe to get missing values in the zip code column
location_df= crash_df_cleaned.filter(["LATITUDE", "LONGITUDE", "ZIP CODE"], axis =1)
location_df=location_df.reset_index(drop=True)
location_df
```

```python
# getting locations only for missing zipcodes
location_only_nozipcode = location_df[location_df["ZIP CODE"].isnull()
```

```python
# dropping row with missing values across in all three column
location_df_nonna = location_only_nozipcode.dropna(axis=0, how='all')
location_df_nonna
```

```python
# save the dataframe location_df_nonna as csv
location_df_nonna.to_csv("location_missingzipcode.csv")
```

```python
# Dropping ZIP CODE column to prepare the dataframe for geopy to get ZIP CODE
location_getZipCode = location_df_nonna.drop(["ZIP CODE"], axis=1)
location_getZipCode_nozero= location_getZipCode.loc[(location_getZipCode!=0).any(axis=1)]  # dropping rows with zero values in location by keeping the rows with values
print(location_getZipCode_nozero.shape)
```

```python
# import geopy to get zip code by using locations
from geopy.geocoders import Nominatim
```

```python
# impport GeocoderTimeOut to get zipcode to avoid timeout error
from geopy.exc import GeocoderTimedOut
```

```python
# get zip code with location with geopy with conditions such as pass the key error and timeout 
import geopy
geolocator = geopy.Nominatim(user_agent="123456")
def get_zipcode(df, geolocator, lat_field, lon_field, attempt=1, max_attempts=100):
    try:
        location = geolocator.reverse((df[lat_field], df[lon_field]))
        return location.raw['address']['postcode']
    except KeyError:
        pass
    except GeocoderTimedOut:
        if attempt <= max_attempts:
            return get_zipcode (df, attempt=attempt+1)
        raise 
```

```python
# get the zipcode based on the location, note that this code takes about 30min 
zipcodes1 = location_getZipCode_nozero.apply(get_zipcode, axis=1, geolocator=geolocator, lat_field='LATITUDE', lon_field='LONGITUDE')
```

```python
# check the zipcodes extracted by geopy based on the given location
zipcodes1
```

```python
# Make a simple dataframe with extracted zipcodes keeping the index 
ZIP_CODE_df =pd.DataFrame({"Missing_ZIP CODE":zipcodes1})
ZIP_CODE_df
#ZIP_CODE_df.info()
```

```python
# merge the zipcode dataframe into the location_df_nonna
Location_zipcode = pd.merge(location_df_nonna,ZIP_CODE_df, left_index=True, right_index=True)
```

```python
# after merge drop "ZIP CODE", "Missing_ZIP CODE_y" to have 3 columns in the dataframe
location_zipcode_full = Location_zipcode.drop(["ZIP CODE", "Missing_ZIP CODE_y"], axis=1)
location_zipcode_full
```

```python
# rename the column name consistant with original dataframe
location_zipcode_full.columns = ["LATITUDE","LONGITUDE","ZIP CODE"]
```

```python
# there is 2 rows still missing with zipcode, so drop those 2 rows
location_zipcode_full=location_zipcode_full.dropna()
```

```python
# save this dataframe as vsc 
#location_zipcode_full.to_csv("zipcode.csv")
location_zipcode_full
```

```python
# change zip codes values from object to integer to be able to pass into the next cell for the conditional statement
location_zipcode_full['ZIP CODE'] = location_zipcode_full['ZIP CODE'].astype(str).astype(int)
location_zipcode_full.info()
```

```python
# using follwoing zip code range for boroughs, make a empty BOROUGH column and fill it with conditional statement with zip code
# Manhattan: 10001-10282
# Staten Island: 10301-10314
# Bronx: 10451-10475
# Queens: 11004-11109, 11351-11697
# Brooklyn: 11201-11256
location_zipcode_full["BOROUGH"]=""
location_zipcode_full.loc[(location_zipcode_full["ZIP CODE"] > 10000) & (location_zipcode_full["ZIP CODE"] < 10283 ), "BOROUGH"] = "MANHATTAN"
location_zipcode_full.loc[(location_zipcode_full["ZIP CODE"] > 10300) & (location_zipcode_full["ZIP CODE"] < 10315 ), "BOROUGH"] = "STATEN ISLAND"
location_zipcode_full.loc[(location_zipcode_full["ZIP CODE"] > 10450) & (location_zipcode_full["ZIP CODE"] < 10476 ), "BOROUGH"] = "BRONX"
location_zipcode_full.loc[(location_zipcode_full["ZIP CODE"] > 11003) & (location_zipcode_full["ZIP CODE"] < 11110 ), "BOROUGH"] = "QUEENS"
location_zipcode_full.loc[(location_zipcode_full["ZIP CODE"] > 11350) & (location_zipcode_full["ZIP CODE"] < 11698 ), "BOROUGH"] = "QUEENS"
location_zipcode_full.loc[(location_zipcode_full["ZIP CODE"] > 11200) & (location_zipcode_full["ZIP CODE"] < 11257 ), "BOROUGH"] = "BROOKLYN"
```

```python
# check the dataframe after filled BOROUGH column with names of boroughs based on zip code
location_zipcode_full
```

```python
# fill the missing values into the original dataframe with the extracted datarame values for zip code and borough
crash_df_cleaned.loc[location_zipcode_full.index, "BOROUGH"] = location_zipcode_full["BOROUGH"]
crash_df_cleaned.loc[location_zipcode_full.index, "ZIP CODE"] = location_zipcode_full["ZIP CODE"]
crash_df_cleaned.head(10)
```

```python
# drop nan rows from the dataframe, now we get 9193 rows which is 3000 more than dropping nan before getting zipcode and borough
crash_nona_df = crash_df_cleaned.dropna()
print(crash_nona_df.shape)
crash_nona_df
```

```python
# set index with COLLISION_ID
crash_nona_index_df=crash_nona_df.reset_index().set_index('COLLISION_ID')
```

```python
# format the CRASH DATE from m/d/y to t-m-d to be able to sort 
crash_nona_index_df['CRASH DATE'] = pd.to_datetime(crash_nona_index_df['CRASH DATE'], format='%m/%d/%y').dt.strftime('%Y-%m-%d')
```

```python
# fromat crash time to be able to sort by time in the next cell 
crash_nona_index_df['CRASH TIME'] = pd.to_datetime(crash_nona_index_df['CRASH TIME'], format= '%H:%M') - pd.to_datetime(crash_nona_index_df['CRASH TIME'], format= '%H:%M').dt.normalize()
```

```python
# sort the dataframe with CRASH DATE and CRASH TIME 
crash_date_sorted_df = crash_nona_index_df.sort_values(['CRASH DATE', "CRASH TIME"] , ascending=[True, True])
crash_date_sorted_df.head()
```

```python
crash_date_sorted_df['CRASH TIME'] = crash_date_sorted_df['CRASH TIME'].astype(str).str.replace("0 days ", "")
crash_date_sorted_df.head()
```

```python
# format zipcode to integer to remove the decimal number
crash_date_sorted_df['ZIP CODE'] = crash_date_sorted_df['ZIP CODE'].astype(int)
crash_date_sorted_df
```
